### test 데이터에 대해서 neutral or dissatisfied라고 예측할 확률을 구하고 그 확률 값을 제출하라

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/x_test.csv")


display(x_train.head())
display(y_train.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

y = y_train['satisfaction']

x_train['Arrival Delay in Minutes'] = x_train['Arrival Delay in Minutes'].fillna(x_train['Arrival Delay in Minutes'].mean())
x_test['Arrival Delay in Minutes'] = x_test['Arrival Delay in Minutes'].fillna(x_test['Arrival Delay in Minutes'].mean())
# print(x_train.info())

x_train_drop = x_train.drop(columns = ['ID', 'id'])
x_test_drop = x_test.drop(columns = ['ID', 'id'])

x_trian_dum = pd.get_dummies(x_train_drop)
x_test_dum = pd.get_dummies(x_test_drop)

scaler = MinMaxScaler()
x_train_scaler = scaler.fit_transform(x_trian_dum)
x_test_scaler = scaler.transform(x_test_dum)
x_train_scaler = pd.DataFrame(x_train_scaler)
x_test_scaler = pd.DataFrame(x_test_scaler)

X_train, X_val, Y_train, Y_val = train_test_split(x_train_scaler, y, test_size = 0.3, random_state = 1)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(500, max_depth = 10, random_state = 1)
rf.fit(X_train, Y_train)

# print(rf.classes_)
# rf.classes_로 ['neutral or dissatisfied', 'satisfied']와 같이 0번째 레이블, 1번째 레이블 확인 가능
# 'neutral or dissatisfied'의 경우 0번째 -> 'neutral or dissatisfied'확률 : train_proba[:,0]

train_label = rf.predict(X_train)
train_proba = rf.predict_proba(X_train)[:,1]

val_label = rf.predict(X_val)
val_proba = rf.predict_proba(X_val)[:,1]

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
print('accuracy_score :', accuracy_score(train_label, Y_train), accuracy_score(val_label, Y_val))
print('f1_score :', f1_score(train_label, Y_train, pos_label = 'neutral or dissatisfied'), f1_score(val_label, Y_val, pos_label = 'neutral or dissatisfied'))
print('precision_score :', precision_score(train_label, Y_train, pos_label = 'neutral or dissatisfied'), precision_score(val_label, Y_val, pos_label = 'neutral or dissatisfied'))
print('recall_score :', recall_score(train_label, Y_train, pos_label = 'neutral or dissatisfied'), recall_score(val_label, Y_val, pos_label = 'neutral or dissatisfied'))
# print('roc_auc_score :', roc_auc_score(train_proba, Y_train), roc_auc_score(val_proba, Y_val))

test_proba = rf.predict_proba(x_test_scaler)[:,0]

df = pd.DataFrame({'ID':x_test['ID'], 'satisfaction':test_proba})
df.to_csv('00000000.csv', index = False)

```
