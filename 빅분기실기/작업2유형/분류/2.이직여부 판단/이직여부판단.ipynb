```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/HRdata/X_test.csv")


display(x_train.head())
display(y_train.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier


drop_col = ['enrollee_id', 'city']

x_train_drop = x_train.drop(columns = drop_col)
x_test_drop = x_test.drop(columns = drop_col)

x_train_dum = pd.get_dummies(x_train_drop)
x_test_dum = pd.get_dummies(x_test_drop)
y = y_train['target']

x_test_dum = x_test_dum[x_train_dum.columns]
# print(y_train)
# print(x_test_dum)


X_train, X_val, Y_train, Y_val = train_test_split(x_train_dum, y, test_size = 0.3, random_state = 1) 

rf = RandomForestClassifier(1000, max_depth = 12, random_state = 3)
rf.fit(X_train, Y_train)

# print(x_train_dum.columns)
# print(x_test_dum.columns)

train_lable = rf.predict(X_train)
train_proba = rf.predict_proba(X_train)[:,1]

val_lable = rf.predict(X_val)
val_proba = rf.predict_proba(X_val)[:,1]
# val_label = rf.predict(x_val

from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
# print(dir(sklearn.metrics))

print('accuracy_score :', accuracy_score(Y_train, train_lable), accuracy_score(Y_val, val_lable))
print('f1_score :', f1_score(Y_train, train_lable), f1_score(Y_val, val_lable))
print('precision_score :', precision_score(Y_train, train_lable), precision_score(Y_val, val_lable))
print('recall_score :', recall_score(Y_train, train_lable), recall_score(Y_val, val_lable))


test_lable = rf.predict(x_test_dum)

# print(test_lable)

df = pd.DataFrame({'enrollee_id' : x_test['enrollee_id'], 'target' : test_lable})
df.to_csv("0000000.csv", index = False)
```
