## 정규성, 등분산성 검정
- 주어진 데이터프레임의 class A,B의 점수가 정규성, 등분산성을 만족하는가? 

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/scipy/scipy2.csv')
```

```python
print(df.head())
print(df.info())
```

```python
dfa = df[df['class'] == 'A']
dfb = df[df['class'] == 'B']
```

```python
from scipy.stats import shapiro

# 귀무가설 : 정규성을 만족한다.
# 대립가설 : 정규성을 만족하지 않음

print(shapiro(dfa['score']))
print(shapiro(dfb['score']))

# A, B 모두 p_value 값이 0.05이하이므로 귀무가설 기각 
# 정규성을 만족하지 않는다.
```

- 등분산성 검정

```python
import scipy.stats
print(dir(scipy.stats))
```

```python
# bartlett : 정규성을 충족하는 2개집단
# fligner : 정규성을 충족하지 않는 2개이상 비모수 집단
#   - 정규성 충족 : center = 'mean'
#   - 정규성 미충족 : center = 'median'
# levene : fligner와 같음, 비모수 데이터에 더 강한 검정력

from scipy.stats import bartlett 
from scipy.stats import fligner
from scipy.stats import levene

a = bartlett(dfa['score'], dfb['score'])
b = fligner(dfa['score'], dfb['score'], center = 'mean')
c = fligner(dfa['score'], dfb['score'], center = 'median')
d = levene(dfa['score'], dfb['score'], center = 'mean')
e = levene(dfa['score'], dfb['score'], center = 'median')
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
from scipy.stats import bartlett, fligner, levene

print(bartlett(dfa['score'], dfb['score']))
print(fligner(dfa['score'], dfb['score'], center = 'median'))
print(levene(dfa['score'], dfb['score'], center = 'median'))
```

## chi square 적합성 검정
- 범주형 변수의 비율이 내가 생각한 비율과 일치하는지 검정

### 특정비율 검정

```python
data = {'blood_type': ['A']*16 + ['B']*16 + ['O']*4 + ['AB']*4}
data = pd.DataFrame(data)

#####################이 부분 중요######################################
data['blood_type'] = pd.Categorical(data['blood_type'],
                                    categories=['A', 'B', 'O', 'AB'])
#print(data)
observed = data['blood_type'].value_counts().sort_index()
#######################################################################

print(observed.to_list())
```

```python
# [1] 혈액형의 비율이 A:B:O:AB = 2:2:3:3 일 것이다. 
import numpy as np
from scipy.stats import chisquare

observed = [16,16,4,4] # 실제 관찰값
expected = sum(observed) * np.array([0.2, 0.2, 0.3, 0.3])

chisquare(observed, expected)
# p_value 0.05 미만 : 귀무가설 기각
# 예상 비율과 일치하지 않는다
```

```python
# [2] 관찰값의 비율이 동일하다면

observed = [54, 46, 60, 40]
expected = sum(observed) * np.array([0.25]*4)

print(chisquare(observed, expected))
# p_value 0.05 이상 : 귀무가설 채택
# 예상 비율과 일치한다.
```

## 윌콕슨 중앙값 검정
- T검정에서 정규성을 만족하지 않을때 평균이 아닌 중앙값으로 검정

### 1samp wilcoxon

```python
# 회사의 닭 가슴살 중량 데이터
df = pd.read_csv('data_02/chicken_breast.csv')
print(df)
```

```python
from scipy.stats import shapiro

print(shapiro(df['weight']))
# p_value 0.05 이상 : 귀무가설 채택...
# 생산된 닭가슴살의 중량은 정규분포를 따른다.

# 정규분포를 따르지만 안따른다고 가정하고..
# T검정이 아닌 윌콕슨검정 실행
```

```python
# [1] 회사에서 생산 한 닭가슴살의 중량의 중앙값은 100g일 것이다.
print(df['weight'].median())

from scipy.stats import wilcoxon
print(wilcoxon(df['weight'] - 100, alternative = 'two-sided'))
# df['weight'] - 100 : (측정값 - 예상 중앙값)

# p_value 0.05 이상 : 귀무가설 채택
# 닭가슴살의 중량의 중앙값은 100g이다
```

```python
# [2] 회사에서 생산 한 닭가슴살의 중량의 중앙값은 100g보다 크거나 같다.

from scipy.stats import wilcoxon
print(wilcoxon(df['weight'] - 100, alternative = 'less'))
# df['weight'] - 100 : (측정값 - 예상 중앙값)

# p_value 0.05 이상 : 귀무가설 채택
# 닭가슴살의 중량의 중앙값은 100g보다 크거나 같다
```

```python
# [2] 회사에서 생산 한 닭가슴살의 중량의 중앙값은 100g보다 작거나 같다.

from scipy.stats import wilcoxon
print(wilcoxon(df['weight'] - 100, alternative = 'greater'))
# df['weight'] - 100 : (측정값 - 예상 중앙값)

# p_value 0.05 이상 : 귀무가설 채택
# 닭가슴살의 중량의 중앙값은 100g보다 작거나 같다
```

```python
# [3] before와 after의 차이가 0 이하이다. (before - after <= 0)
from scipy.stats import wilcoxon

print(wilcoxon(df['before'], df['after'], alternative = 'greater'))
# p_value 0.05 이상 : 귀무가설 기각
# before, after의 차이는 0 이하가 아니다.
```

## 독립2표본 중앙값 점검

```python
# A사, B사의 닭가슴살 중량 데이터
import pandas as pd
df = pd.read_csv('data_02/chicken_weight_AB.csv')
df.tail()
```

```python
gA = df[df['company'] == 'A']['weight']
gB = df[df['company'] == 'B']['weight']
```

```python
# wilcoxon ranksum 검정 (두 독립집단)

# [1] 두 회사의 닭가슴살 중량은 같다.
from scipy.stats import ranksums
print(ranksums(gA,gB, alternative = 'two-sided'))
# pvalue 0.05 미만, 귀무가설 기각 : 중량이 같지 않다

# [2] A-B의 닭가슴살 중량은 0보다 크거나 같다.(A가 더 크다)
print(ranksums(gA,gB, alternative = 'less'))
# pvalue 0.05 미만, 귀무가설 기각 : A-B는 0보다 작다 (B가 더 크다)

# [3] A-B의 닭가슴살 중량은 0보다 작거나 같다.(B가 더 크다)
print(ranksums(gA,gB, alternative = 'greater'))
# pvalue 0.05 이상, 귀무가설 채택 : A-B는 0보다 작다 (B가 더 크다)
```
