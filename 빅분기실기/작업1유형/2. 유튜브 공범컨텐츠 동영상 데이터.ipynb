18. video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고 videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라

```python
video_ratio = video.sort_values(by = 'ct', ascending = False).drop_duplicates('videoname')
video_ratio['ratio'] = (video_ratio['dislikecnt'] / video_ratio['likecnt'])
video_ratio[['videoname', 'ratio']].sort_values(by = 'ratio')
```

19. ????2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오

```python
video00 = video[video.ct.dt.date == pd.to_datetime('2021-11-01')].drop_duplicates('videoname')[['videoname', 'viewcnt']].reset_index(drop = True).sort_values('videoname')
video15 = video[video.ct.dt.date == pd.to_datetime('2021-11-01')].sort_values('ct', ascending = False).drop_duplicates('videoname')[['videoname', 'viewcnt']].reset_index(drop = True).sort_values('videoname')
# display(video00, video15)
# video15['viewcnt'] - video00['viewcnt']
```

```python
start = pd.to_datetime("2021-11-01 00:00:00")
end = pd.to_datetime("2021-11-01 15:00:00")

target = video.loc[(video["ct"] >= start) & (video['ct'] <= end)].reset_index(drop=True)

def check(x):
    result = max(x) - min(x)
    return result

answer = target[['videoname','viewcnt']].groupby("videoname").agg(check)
answer
```
